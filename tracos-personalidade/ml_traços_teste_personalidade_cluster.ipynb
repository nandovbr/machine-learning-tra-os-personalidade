#### Instalando e instanciando as bibliotecas para uso

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
# para teste de visualização dos dados com número ideal de clusters
```

```python
# para teste de visualização dos dados com número ideal de clusters
kmeans = KMeans()
visualizer = KElbowVisualizer(kmeans, k=(2, 10))
 # k=(2, 10) é o número de grupos que o kmeans fará o agrupamento dos dados para teste a partir de 2 grupos até 10 grupos
```

#### Selecionando uma amostra aleatória dos dados com 5000 observações e executando o teste

```python
data_sample = data.sample(n=5000, random_state=1)
# para teste de visualização dos dados com número ideal de clusters
```

```python
visualizer.fit(data_sample) # fit() é o método que executa os parametros do kmeans
visualizer.poof() # poof() é o método que exibe o gráfico com o número ideal de clusters
# para teste de visualização dos dados com número ideal de clusters
```

* Nesse resultado acima, o número ideal de Clusters são 5

#### Agrupando os participantes que responderam aos testes em 5 grupos

```python
kmeans = KMeans(n_clusters=5) # n_clusters=5 é o número de grupos que o kmeans fará o agrupamento dos dados
k_fit = kmeans.fit(data) # fit() é o método que executa os parametros do kmeans
```

```python
predicoes = k_fit.labels_ # labels_ é o método que retorna o resultado do agrupamento dos dados
data['Clusters'] = predicoes # adiciona a coluna Clusters com os resultados do agrupamento dos dados
```

```python
data.head()
# verificando se a coluna Clusters foi inserida com sucesso
```

#### Analisando os grupos

```python
data['Clusters'].value_counts()
# quantidade de observações de cada cluster em ordem crescente
```

```python
# Agrupando os registros por Clusters e pegando a media
data.groupby('Clusters').mean()
```

```python
# Calculando a média de cada grupo de questões para verificar um padrão de respostas
# selecionando as colunas de cada grupo
col_list = list(data)
ext = col_list[0:10] # variável que armazena as colunas de 0 a 10
est = col_list[10:20] # variável que armazena as colunas de 10 a 20
agr = col_list[20:30] # variável que armazena as colunas de 20 a 30
csn = col_list[30:40] # variável que armazena as colunas de 30 a 40
opn = col_list[40:50] # variável que armazena as colunas de 40 a 50
```

```python
# Criando um novo dataframe para armazenar as médias de cada grupo criado na célula acima
data_soma = pd.DataFrame()
data_soma['extroversion'] = data[ext].sum(axis=1)/10
data_soma['neurotic'] = data[est].sum(axis=1)/10
data_soma['agreeable'] = data[agr].sum(axis=1)/10
data_soma['conscientious'] = data[csn].sum(axis=1)/10
data_soma['open'] = data[opn].sum(axis=1)/10
data_soma['Clusters'] = predicoes # adiciona a coluna Clusters com os resultados do agrupamento dos dados
```

```python
# valor médio por cluster
data_soma.groupby('Clusters').mean()
```

#### Visualizar as medias por grupo e plotagem dos gráficos

```python
data_clusters = data_soma.groupby('Clusters').mean() # salva o resultado do agrupamento dos dados em uma variável
```

```python
# plotando o gráfico de barras para visualizar o resultado do agrupamento e linhas para visualizar o padrão de respostas
plt.figure(figsize=(22,3))
for i in range(0, 5):
    plt.subplot(1, 5, i+1)
    plt.bar(data_clusters.columns, data_clusters.iloc[:, i], color='green', alpha=0.2)
    plt.plot(data_clusters.columns, data_clusters.iloc[:, i], color='red')
    plt.title('Cluster ' + str(i+1))
    plt.xticks(rotation=45)
    plt.ylim(0, 4);
```

```python
import gradio as gr
```

```python
# Lendo os dados com as questões e colocando cada um em uma linha do dataframe
dicio_questions = open('questions.txt').read().split('\n')
```

```python
dicio_questions
```

```python
# limpando os dados  e recuperando apenas as questões
questions = []
for q in dicio_questions:
    q = str(q) # transformar cada linha em string
    questions.append(q[q.find('\t'):].lstrip()) # ler as strings depois do \t e remove os espaços em branco antes e depois da questão
```

```python
questions
```

```python
# Criando inputs dinamicos para passar ao gradio
input_questions = []
for q in questions:
    obj_input = gr.inputs.Slider(minimum=1,maximum=5,step=1,default=3,label=q)
    input_questions.append(obj_input)
```

```python
input_questions # virificando os inputs
```

#### Criando a interface e a função predict

```python
def predict(*outputs_questions):
    outputs_questions = np.array(outputs_questions).reshape(1, -1)
    return k_fit.predict(outputs_questions)

iface = gr.Interface(
                    fn = predict, # função que será executada
                    title = 'Big Five Personality', # título da interface
                    description = 'Sistema para detecção de traços de personalidade', # descrição da interface
                    inputs = input_questions, # inputs da interface
                    outputs='text',) # outputs da interface com o cluster ao qual o usuário pertencerá após envio das respostas
iface.launch(share=True) # gera um IP publico q posso compartilhar com outras pessoas
```
